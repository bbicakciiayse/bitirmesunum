```python

# ==============================
# IMPORTS
# ==============================
import warnings
from pathlib import Path
from typing import Dict, List, Optional, Tuple
import textwrap

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

from sklearn.compose import ColumnTransformer
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    f1_score,
    precision_score,
    recall_score,
    roc_auc_score,
)
from sklearn.model_selection import (
    GridSearchCV,
    StratifiedKFold,
    cross_val_score,
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)


# ==============================
# CONFIGURATION
# ==============================
BASE_DIR = Path.cwd()
ARTIFACTS_DIR = BASE_DIR / "artifacts"
PLOTS_DIR = ARTIFACTS_DIR / "plots"
OUTPUTS_DIR = ARTIFACTS_DIR / "outputs"

# Kendi dosya yolunu burada güncelle
EXCEL_FILE = Path(r"C:\Users\ayseb\OneDrive\Desktop\Bitirme Offers copy.xlsx")

TARGET_COLUMN = "Result"
PRICE_COLUMN = "Price"
OPPORTUNITY_TYPE_COLUMN = "Opportunity Type"
USER_COLUMN = "User"

RANDOM_STATE = 42
CV_FOLDS = 5
NESTED_OUTER_FOLDS = 3
NESTED_INNER_FOLDS = 3
N_JOBS = 1

MANUAL_ENGINEERED_FEATURES = [
    "IAM, Cybersecurity or IT Department",
    "Company Size",
]

LEAKAGE_RISK_COLUMNS = [
    "Status", "Final Status", "Outcome", "Won Date", "Lost Date",
    "Close Date", "Closed Date", "Decision", "Win Probability", "Win Rate"
]

LEAKAGE_KEYWORDS = [
    "result"
]



for path in [ARTIFACTS_DIR, PLOTS_DIR, OUTPUTS_DIR]:
    path.mkdir(parents=True, exist_ok=True)

print("Artifacts directory:", ARTIFACTS_DIR.resolve())
print("Outputs directory:", OUTPUTS_DIR.resolve())
print("Excel file path:", EXCEL_FILE)
print("Excel exists:", EXCEL_FILE.exists())


# ==============================
# COMMON HELPERS
# ==============================
def find_column_ignore_case(columns, name):
    for col in columns:
        if str(col).lower() == str(name).lower():
            return col
    return None


def standardize_column_names(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    rename_map = {}

    typo_col = find_column_ignore_case(df.columns, "Oportunity Type")
    if typo_col is not None and typo_col != OPPORTUNITY_TYPE_COLUMN:
        rename_map[typo_col] = OPPORTUNITY_TYPE_COLUMN

    quoted_price_col = find_column_ignore_case(df.columns, "Quoted Price")
    if quoted_price_col is not None and quoted_price_col != PRICE_COLUMN:
        rename_map[quoted_price_col] = PRICE_COLUMN

    hash_user_col = find_column_ignore_case(df.columns, "#user")
    if hash_user_col is not None and hash_user_col != USER_COLUMN:
        rename_map[hash_user_col] = USER_COLUMN

    if rename_map:
        df = df.rename(columns=rename_map)
        print("Renamed columns:", rename_map)

    return df


def map_target(y_series: pd.Series) -> pd.Series:
    mapping = {
        "won": 1, "win": 1, "yes": 1, "1": 1, "true": 1,
        "lost": 0, "loss": 0, "no": 0, "0": 0, "false": 0,
    }
    return y_series.astype(str).str.strip().str.lower().map(mapping)


def show_df(df: pd.DataFrame, title: Optional[str] = None, n: int = 10):
    if title:
        print(f"\n{title}")
    display(df.head(n))


def find_suspicious_columns(columns: List[str]) -> List[str]:
    suspicious = []
    for col in columns:
        norm = str(col).strip().lower().replace("_", " ")
        if any(keyword in norm for keyword in LEAKAGE_KEYWORDS):
            suspicious.append(col)
    return suspicious


def get_feature_names_from_pipeline(pipeline: Pipeline) -> List[str]:
    preprocessor = pipeline.named_steps["preprocessor"]
    feature_names = []

    for name, transformer, cols in preprocessor.transformers_:
        if transformer == "drop" or not cols:
            continue

        if name == "num":
            feature_names.extend(cols)

        elif name == "cat":
            encoder = transformer.named_steps["encoder"]
            try:
                encoded_names = encoder.get_feature_names_out(cols)
                feature_names.extend(encoded_names.tolist())
            except Exception:
                feature_names.extend(cols)

    return feature_names


# ==============================
# SAFE ONEHOTENCODER FOR DIFFERENT SKLEARN VERSIONS
# ==============================
def make_onehot_encoder():
    try:
        return OneHotEncoder(
            handle_unknown="ignore",
            drop="first",
            sparse_output=False
        )
    except TypeError:
        return OneHotEncoder(
            handle_unknown="ignore",
            drop="first",
            sparse=False
        )


# ==============================
# FEATURE ENGINEERING
# ==============================
def add_feature_engineering(df: pd.DataFrame) -> Tuple[pd.DataFrame, List[str]]:
    df = df.copy()
    engineered_columns = []

    offer_date_col = find_column_ignore_case(df.columns, "Offer Date")
    if offer_date_col is not None:
        offer_dates = pd.to_datetime(df[offer_date_col], errors="coerce")
        df["Offer_Year"] = offer_dates.dt.year
        df["Offer_Month"] = offer_dates.dt.month
        df["Offer_Quarter"] = offer_dates.dt.quarter
        engineered_columns.extend(["Offer_Year", "Offer_Month", "Offer_Quarter"])
        df = df.drop(columns=[offer_date_col])

    price_col = find_column_ignore_case(df.columns, PRICE_COLUMN)
    if price_col is not None:
        price_numeric = pd.to_numeric(df[price_col], errors="coerce")
        df["Price_Log"] = np.log1p(price_numeric.clip(lower=0))
        engineered_columns.append("Price_Log")
    else:
        price_numeric = None

    user_col = find_column_ignore_case(df.columns, USER_COLUMN)
    if user_col is not None:
        user_numeric = pd.to_numeric(df[user_col], errors="coerce")
        df["User_Log"] = np.log1p(user_numeric.clip(lower=0))
        engineered_columns.append("User_Log")

        if price_numeric is not None:
            safe_users = user_numeric.replace(0, np.nan)
            df["Price_Per_User"] = price_numeric / safe_users
            df["Price_Per_User"] = df["Price_Per_User"].replace([np.inf, -np.inf], np.nan)
            df["Price_Per_User_Log"] = np.log1p(df["Price_Per_User"].clip(lower=0))
            engineered_columns.extend(["Price_Per_User", "Price_Per_User_Log"])

    competition_col = find_column_ignore_case(df.columns, "Competition")
    if competition_col is not None:
        competition_text = df[competition_col].fillna("").astype(str).str.strip()
        df["Has_Competition"] = (
            (competition_text != "") &
            (competition_text.str.lower() != "nan")
        ).astype(float)

        df["Competition_Count"] = competition_text.apply(
            lambda x: len([part for part in x.replace(";", ",").split(",") if part.strip()])
            if x and x.lower() != "nan" else 0
        )
        engineered_columns.extend(["Has_Competition", "Competition_Count"])

    partner_col = find_column_ignore_case(df.columns, "Partner")
    if partner_col is not None:
        partner_text = df[partner_col].fillna("").astype(str).str.strip()
        df["Has_Partner"] = (
            (partner_text != "") &
            (partner_text.str.lower() != "nan")
        ).astype(float)
        engineered_columns.append("Has_Partner")

    product_col = find_column_ignore_case(df.columns, "Product")
    opp_type_col = find_column_ignore_case(df.columns, OPPORTUNITY_TYPE_COLUMN)
    if product_col is not None and opp_type_col is not None:
        df["Product_Opportunity_Type"] = (
            df[product_col].fillna("missing").astype(str)
            + " | "
            + df[opp_type_col].fillna("missing").astype(str)
        )
        engineered_columns.append("Product_Opportunity_Type")

    sector_col = find_column_ignore_case(df.columns, "Sector")
    source_col = find_column_ignore_case(df.columns, "Opportunity Source")
    if sector_col is not None and source_col is not None:
        df["Sector_Source"] = (
            df[sector_col].fillna("missing").astype(str)
            + " | "
            + df[source_col].fillna("missing").astype(str)
        )
        engineered_columns.append("Sector_Source")

    project_end_col = find_column_ignore_case(df.columns, "Estimated Project End Quarter")
    if project_end_col is not None:
        project_end = df[project_end_col].astype(str).str.strip()
        df["Project_End_Known"] = (
            (project_end != "") &
            (project_end.str.lower() != "nan")
        ).astype(float)
        engineered_columns.append("Project_End_Known")

    return df, engineered_columns


# ==============================
# DATA LOADING
# ==============================
def load_dataset(
    excel_file: Path = EXCEL_FILE,
    use_feature_engineering: bool = True,
    drop_leakage_risk: bool = False,
):
    if not Path(excel_file).exists():
        raise FileNotFoundError(
            f"Excel file not found: {excel_file}. Lütfen EXCEL_FILE değişkenini kendi dosyana göre güncelle."
        )

    df = pd.read_excel(excel_file)
    df = standardize_column_names(df)
    print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")

    target_col = find_column_ignore_case(df.columns, TARGET_COLUMN) or TARGET_COLUMN
    if target_col not in df.columns:
        raise ValueError(f"Target column '{TARGET_COLUMN}' not found.")

    y = map_target(df[target_col])
    valid_mask = ~y.isna()

    if not valid_mask.all():
        removed = int((~valid_mask).sum())
        print(f"Removed {removed} rows with invalid target values.")
        df = df.loc[valid_mask].copy()
        y = y.loc[valid_mask].copy()

    y = y.astype(int)
    X = df.drop(columns=[target_col]).copy()

    removed_columns = []
    if drop_leakage_risk:
        removable = {col for col in LEAKAGE_RISK_COLUMNS if col in X.columns}
        removable.update(find_suspicious_columns(list(X.columns)))
        removed_columns = sorted(removable)

        if removed_columns:
            X = X.drop(columns=removed_columns)
            print("Removed leakage-risk columns:", removed_columns)

    engineered_columns = []
    if use_feature_engineering:
        X, engineered_columns = add_feature_engineering(X)
        if engineered_columns:
            print(f"Engineered feature count: {len(engineered_columns)}")

    numerical_cols = [col for col in X.columns if pd.api.types.is_numeric_dtype(X[col])]
    categorical_cols = [col for col in X.columns if col not in numerical_cols]

    for col in categorical_cols:
        X[col] = X[col].fillna("missing").astype(str)

    price_col = find_column_ignore_case(X.columns, PRICE_COLUMN) or PRICE_COLUMN
    if price_col not in X.columns:
        print(f"Warning: price column '{PRICE_COLUMN}' not found. Price sensitivity may fail.")

    return {
        "X": X,
        "y": y,
        "numerical_cols": numerical_cols,
        "categorical_cols": categorical_cols,
        "price_col": price_col,
        "engineered_columns": engineered_columns,
        "removed_columns": removed_columns,
    }


# ==============================
# PREPROCESSING + MODELS
# ==============================
def create_preprocessor(numerical_cols, categorical_cols, scale_numerical=True):
    transformers = []

    if numerical_cols:
        num_steps = [("imputer", SimpleImputer(strategy="median"))]
        if scale_numerical:
            num_steps.append(("scaler", StandardScaler()))
        transformers.append(
            ("num", Pipeline(num_steps), numerical_cols)
        )

    if categorical_cols:
        transformers.append(
            (
                "cat",
                Pipeline([
                    ("imputer", SimpleImputer(strategy="most_frequent")),
                    ("encoder", make_onehot_encoder()),
                ]),
                categorical_cols
            )
        )

    return ColumnTransformer(transformers=transformers, remainder="drop")


def get_comparison_models_and_grids():
    models = {
        "Logistic Regression": LogisticRegression(
            penalty=None,
            solver="lbfgs",
            random_state=RANDOM_STATE,
            class_weight="balanced",
            max_iter=5000
        ),
        "Ridge Logistic Regression": LogisticRegression(
            penalty="l2",
            solver="lbfgs",
            random_state=RANDOM_STATE,
            class_weight="balanced",
            max_iter=5000
        ),
        "Lasso Logistic Regression": LogisticRegression(
            penalty="l1",
            solver="liblinear",
            random_state=RANDOM_STATE,
            class_weight="balanced",
            max_iter=5000
        ),
        "Elastic Net Logistic Regression": LogisticRegression(
            penalty="elasticnet",
            solver="saga",
            random_state=RANDOM_STATE,
            class_weight="balanced",
            max_iter=5000
        ),
        "Random Forest": RandomForestClassifier(
            random_state=RANDOM_STATE,
            class_weight="balanced"
        ),
        "Gradient Boosting": GradientBoostingClassifier(
            random_state=RANDOM_STATE
        ),
    }

    param_grids = {
        "Logistic Regression": {},
        "Ridge Logistic Regression": {
            "classifier__C": [0.01, 0.1, 1, 10, 100]
        },
        "Lasso Logistic Regression": {
            "classifier__C": [0.01, 0.1, 1, 10, 100]
        },
        "Elastic Net Logistic Regression": {
            "classifier__C": [0.01, 0.1, 1, 10, 100],
            "classifier__l1_ratio": [0.1, 0.5, 0.9]
        },
        "Random Forest": {
            "classifier__n_estimators": [100, 200],
            "classifier__max_depth": [None, 10, 20]
        },
        "Gradient Boosting": {
            "classifier__n_estimators": [100, 200],
            "classifier__learning_rate": [0.01, 0.1],
            "classifier__max_depth": [3, 5]
        },
    }

    return models, param_grids


# ==============================
# FEATURE TABLES
# ==============================
def build_engineered_feature_table(data_dict):
    X = data_dict["X"].copy()
    _, manual_engineered, generated_engineered = classify_features(data_dict)
    engineered_cols = manual_engineered + generated_engineered
    removed_cols = data_dict.get("removed_columns", [])

    engineered_df = pd.DataFrame({
        "Feature": engineered_cols,
        "Feature_Source": [
            "MANUAL_ENGINEERED" if col in manual_engineered else "GENERATED_ENGINEERED"
            for col in engineered_cols
        ],
        "Exists_in_X": [col in X.columns for col in engineered_cols],
        "Data_Type": [str(X[col].dtype) if col in X.columns else "missing" for col in engineered_cols],
        "Unique_Values": [X[col].nunique(dropna=False) if col in X.columns else np.nan for col in engineered_cols],
        "Missing_Count": [X[col].isna().sum() if col in X.columns else np.nan for col in engineered_cols],
        "Sample_Values": [
            ", ".join(map(str, X[col].dropna().astype(str).head(3).tolist())) if col in X.columns else ""
            for col in engineered_cols
        ]
    })

    removed_df = pd.DataFrame({
        "Removed_Leakage_Columns": removed_cols
    }) if removed_cols else pd.DataFrame(columns=["Removed_Leakage_Columns"])

    return engineered_df, removed_df


# ==============================
# FEATURE IMPACT ANALYSIS
# ==============================
def get_raw_and_engineered_from_loaded_data(data_dict):
    raw_cols, manual_engineered, generated_engineered = classify_features(data_dict)
    engineered_cols = manual_engineered + generated_engineered
    return raw_cols, engineered_cols


def build_model_for_feature_list(X, feature_list):
    numerical_cols = [c for c in feature_list if pd.api.types.is_numeric_dtype(X[c])]
    categorical_cols = [c for c in feature_list if c not in numerical_cols]

    return Pipeline([
        ("preprocessor", create_preprocessor(numerical_cols, categorical_cols, scale_numerical=True)),
        ("classifier", LogisticRegression(
            penalty="l2",
            solver="lbfgs",
            C=1.0,
            random_state=RANDOM_STATE,
            class_weight="balanced",
            max_iter=5000
        ))
    ])


def evaluate_feature_additions(data_dict):
    X = data_dict["X"].copy()
    y = data_dict["y"].copy()

    raw_cols, manual_engineered, generated_engineered = classify_features(data_dict)
    engineered_cols = manual_engineered + generated_engineered
    cv = StratifiedKFold(n_splits=CV_FOLDS, shuffle=True, random_state=RANDOM_STATE)

    baseline_model = build_model_for_feature_list(X, raw_cols)
    baseline_scores = cross_val_score(
        baseline_model,
        X[raw_cols],
        y,
        cv=cv,
        scoring="roc_auc",
        n_jobs=N_JOBS
    )
    baseline_mean = baseline_scores.mean()

    rows = []
    for feat in engineered_cols:
        test_features = raw_cols + [feat]
        model = build_model_for_feature_list(X, test_features)

        scores = cross_val_score(
            model,
            X[test_features],
            y,
            cv=cv,
            scoring="roc_auc",
            n_jobs=N_JOBS
        )
        mean_score = scores.mean()

        rows.append({
            "Feature": feat,
            "Feature_Source": "MANUAL_ENGINEERED" if feat in manual_engineered else "GENERATED_ENGINEERED",
            "Baseline_RAW_ROC_AUC": baseline_mean,
            "RAW_plus_Feature_ROC_AUC": mean_score,
            "Delta_ROC_AUC": mean_score - baseline_mean,
            "Keep?": "KEEP" if mean_score > baseline_mean else "DROP"
        })

    return pd.DataFrame(rows).sort_values("Delta_ROC_AUC", ascending=False).reset_index(drop=True)


def build_final_feature_set(data_dict, impact_df):
    X = data_dict["X"].copy()
    raw_cols, manual_engineered, generated_engineered = classify_features(data_dict)

    kept_engineered = impact_df.loc[impact_df["Keep?"] == "KEEP", "Feature"].tolist()
    dropped_engineered = impact_df.loc[impact_df["Keep?"] == "DROP", "Feature"].tolist()
    final_features = raw_cols + kept_engineered

    final_df = pd.DataFrame({
        "Feature": final_features,
        "Feature_Type": ["RAW" if c in raw_cols else "KEEP" for c in final_features],
        "Feature_Source": [
            "RAW"
            if c in raw_cols else
            "MANUAL_ENGINEERED"
            if c in manual_engineered else
            "GENERATED_ENGINEERED"
            if c in generated_engineered else
            "UNKNOWN"
            for c in final_features
        ]
    })

    dropped_df = pd.DataFrame({
        "Dropped_Engineered_Features": dropped_engineered,
        "Feature_Source": [
            "MANUAL_ENGINEERED" if c in manual_engineered else "GENERATED_ENGINEERED"
            for c in dropped_engineered
        ]
    })

    return final_features, final_df, dropped_df


# ==============================
# NESTED CV FEATURE SUBSET COMPARISON
# ==============================
def infer_engineered_features(columns: List[str]) -> List[str]:
    engineered = []
    for col in columns:
        normalized = str(col).lower().replace(" ", "_")
        if any(keyword in normalized for keyword in ENGINEERED_FEATURE_KEYWORDS):
            engineered.append(col)
    return engineered

def classify_features(data_dict):
    X = data_dict["X"]
    all_columns = list(X.columns)

    # MANUAL ENGINEERED (senin verdiğin)
    manual_engineered = [
        col for col in MANUAL_ENGINEERED_FEATURES
        if col in all_columns
    ]

    # GENERATED ENGINEERED (kodun ürettikleri)
    generated_engineered = [
        col for col in data_dict["engineered_columns"]
        if col in all_columns and col not in manual_engineered
    ]

    # RAW FEATURES (geri kalan)
    raw_features = [
        col for col in all_columns
        if col not in manual_engineered
        and col not in generated_engineered
    ]

    return raw_features, manual_engineered, generated_engineered

def build_pipeline_for_feature_subset(feature_names: List[str], X: pd.DataFrame, estimator):
    numerical_cols = [c for c in feature_names if pd.api.types.is_numeric_dtype(X[c])]
    categorical_cols = [c for c in feature_names if c not in numerical_cols]

    scale_numerical = not isinstance(estimator, (RandomForestClassifier, GradientBoostingClassifier))
    preprocessor = create_preprocessor(
        numerical_cols,
        categorical_cols,
        scale_numerical=scale_numerical
    )

    return Pipeline([
        ("preprocessor", preprocessor),
        ("classifier", estimator)
    ])


def train_feature_subset_nested_cv(feature_set_name, feature_names, X, y):
    base_models, param_grids = get_comparison_models_and_grids()
    results = {}

    outer_cv = StratifiedKFold(
        n_splits=NESTED_OUTER_FOLDS,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    inner_cv = StratifiedKFold(
        n_splits=NESTED_INNER_FOLDS,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    for name, estimator in base_models.items():
        print(f"\nNested CV running for: {feature_set_name} | {name}")

        outer_fold_metrics = []
        outer_best_params = []
        fold_no = 1

        for train_idx, test_idx in outer_cv.split(X[feature_names], y):
            X_train_fold = X.iloc[train_idx][feature_names].copy()
            X_test_fold = X.iloc[test_idx][feature_names].copy()
            y_train_fold = y.iloc[train_idx].copy()
            y_test_fold = y.iloc[test_idx].copy()

            pipe = build_pipeline_for_feature_subset(feature_names, X, estimator)

            grid = GridSearchCV(
                estimator=pipe,
                param_grid=param_grids.get(name, {}),
                cv=inner_cv,
                scoring="roc_auc",
                n_jobs=N_JOBS,
                verbose=0,
                refit=True
            )

            grid.fit(X_train_fold, y_train_fold)
            best_model = grid.best_estimator_

            y_pred = best_model.predict(X_test_fold)
            y_proba = best_model.predict_proba(X_test_fold)[:, 1]

            outer_fold_metrics.append({
                "Outer Fold": fold_no,
                "ROC-AUC": roc_auc_score(y_test_fold, y_proba),
                "Accuracy": accuracy_score(y_test_fold, y_pred),
                "Precision": precision_score(y_test_fold, y_pred, zero_division=0),
                "Recall": recall_score(y_test_fold, y_pred, zero_division=0),
                "F1-score": f1_score(y_test_fold, y_pred, zero_division=0),
                "Best Params": grid.best_params_,
            })

            outer_best_params.append(str(grid.best_params_))
            fold_no += 1

        fold_df = pd.DataFrame(outer_fold_metrics)

        results[name] = {
            "Feature Set": feature_set_name,
            "Model Name": name,
            "Fold Results": fold_df,
            "Nested CV Mean ROC-AUC": fold_df["ROC-AUC"].mean(),
            "Nested CV Std ROC-AUC": fold_df["ROC-AUC"].std(),
            "Nested CV Mean Accuracy": fold_df["Accuracy"].mean(),
            "Nested CV Mean Precision": fold_df["Precision"].mean(),
            "Nested CV Mean Recall": fold_df["Recall"].mean(),
            "Nested CV Mean F1-score": fold_df["F1-score"].mean(),
            "Best Params Per Fold": outer_best_params,
        }

    return results


def nested_feature_subset_results_to_dataframe(results: Dict[str, dict]) -> pd.DataFrame:
    rows = []
    for model_name, res in results.items():
        rows.append({
            "Feature Set": res["Feature Set"],
            "Model": model_name,
            "Nested CV Mean ROC-AUC": res["Nested CV Mean ROC-AUC"],
            "Nested CV Std ROC-AUC": res["Nested CV Std ROC-AUC"],
            "Nested CV Mean Accuracy": res["Nested CV Mean Accuracy"],
            "Nested CV Mean Precision": res["Nested CV Mean Precision"],
            "Nested CV Mean Recall": res["Nested CV Mean Recall"],
            "Nested CV Mean F1-score": res["Nested CV Mean F1-score"],
            "Best Params Per Fold": " | ".join(res["Best Params Per Fold"]),
        })

    return pd.DataFrame(rows).sort_values(
        ["Nested CV Mean ROC-AUC", "Nested CV Mean F1-score"],
        ascending=False
    ).reset_index(drop=True)


def print_nested_feature_subset_results(results: Dict[str, dict]):
    summary_df = nested_feature_subset_results_to_dataframe(results)

    print("\n" + "=" * 90)
    print("NESTED CV FEATURE SUBSET SUMMARY")
    print("=" * 90)
    print(summary_df.to_string(index=False))

    print("\n" + "=" * 90)
    print("DETAILED OUTER FOLD RESULTS")
    print("=" * 90)

    for model_name, res in results.items():
        print(f"\nFeature Set: {res['Feature Set']}")
        print(f"Model: {model_name}")
        print("-" * 90)
        print(f"Nested CV Mean ROC-AUC : {res['Nested CV Mean ROC-AUC']:.4f}")
        print(f"Nested CV Std ROC-AUC  : {res['Nested CV Std ROC-AUC']:.4f}")
        print(f"Nested CV Mean Accuracy: {res['Nested CV Mean Accuracy']:.4f}")
        print(f"Nested CV Mean Precision: {res['Nested CV Mean Precision']:.4f}")
        print(f"Nested CV Mean Recall   : {res['Nested CV Mean Recall']:.4f}")
        print(f"Nested CV Mean F1-score : {res['Nested CV Mean F1-score']:.4f}")
        print("Best Params Per Fold:")
        for i, params in enumerate(res["Best Params Per Fold"], start=1):
            print(f"  Fold {i}: {params}")
        print("\nOuter Fold Table:")
        print(res["Fold Results"].to_string(index=False))
        print("-" * 90)


def run_feature_subset_nested_cv_comparison(
    X: pd.DataFrame,
    y: pd.Series,
    data_dict=None,
    selected_features: Optional[List[str]] = None
):
    if data_dict is None:
        data_dict = {"X": X, "engineered_columns": []}

    raw_features, manual_engineered, generated_engineered = classify_features(data_dict)
    engineered_features = manual_engineered + generated_engineered
    all_features = list(X.columns)
    selected_features = [col for col in (selected_features or all_features) if col in X.columns]
    selected_raw_features = [col for col in raw_features if col in selected_features]
    selected_manual_engineered = [col for col in manual_engineered if col in selected_features]
    selected_generated_engineered = [col for col in generated_engineered if col in selected_features]
    selected_engineered_features = selected_manual_engineered + selected_generated_engineered

    print("\nFeature groups:")
    print(f"Raw feature count       : {len(selected_raw_features)}")
    print(f"Engineered feature count: {len(selected_engineered_features)}")
    print(f"Selected feature count  : {len(selected_features)}")

    all_results = {}

    if selected_raw_features:
        all_results["Raw Features"] = train_feature_subset_nested_cv(
            feature_set_name="Raw Features",
            feature_names=selected_raw_features,
            X=X,
            y=y
        )

    if selected_manual_engineered:
        all_results["Manual Engineered Features"] = train_feature_subset_nested_cv(
            feature_set_name="Manual Engineered Features",
            feature_names=selected_manual_engineered,
            X=X,
            y=y
        )

    if selected_generated_engineered:
        all_results["Generated Engineered Features"] = train_feature_subset_nested_cv(
            feature_set_name="Generated Engineered Features",
            feature_names=selected_generated_engineered,
            X=X,
            y=y
        )

    if selected_engineered_features:
        all_results["All Engineered Features"] = train_feature_subset_nested_cv(
            feature_set_name="All Engineered Features",
            feature_names=selected_engineered_features,
            X=X,
            y=y
        )

    all_results["Selected Features"] = train_feature_subset_nested_cv(
        feature_set_name="Selected Features",
        feature_names=selected_features,
        X=X,
        y=y
    )

    return all_results
def combine_all_nested_subset_results(all_results: Dict[str, Dict[str, dict]]) -> pd.DataFrame:
    frames = []
    for subset_name, subset_results in all_results.items():
        frames.append(nested_feature_subset_results_to_dataframe(subset_results))

    if not frames:
        return pd.DataFrame()

    return pd.concat(frames, axis=0, ignore_index=True).sort_values(
        ["Nested CV Mean ROC-AUC", "Nested CV Mean F1-score"],
        ascending=False
    ).reset_index(drop=True)


# ==============================
# SAVE HELPERS
# ==============================
def save_dataframe(df: pd.DataFrame, filename: str):
    OUTPUTS_DIR.mkdir(parents=True, exist_ok=True)

    excel_path = OUTPUTS_DIR / f"{filename}.xlsx"
    csv_path = OUTPUTS_DIR / f"{filename}.csv"

    df.to_excel(excel_path, index=False)
    df.to_csv(csv_path, index=False, encoding="utf-8-sig")

    print(f"Saved: {excel_path}")
    print(f"Saved: {csv_path}")


# ==============================
# REPORT PLOTS
# ==============================
def save_plot(fig, filename: str):
    PLOTS_DIR.mkdir(parents=True, exist_ok=True)
    png_path = PLOTS_DIR / f"{filename}.png"
    fig.tight_layout()
    fig.savefig(png_path, dpi=300, bbox_inches="tight")
    plt.close(fig)
    print(f"Saved plot: {png_path}")


def style_axis(ax, title: str, ylabel: str, xlabel: Optional[str] = None):
    ax.set_title(title, fontsize=20, fontweight="bold", pad=18)
    ax.set_ylabel(ylabel, fontsize=13)
    if xlabel:
        ax.set_xlabel(xlabel, fontsize=13)
    ax.grid(axis="y", linestyle="--", alpha=0.25)
    ax.set_axisbelow(True)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)


def wrap_tick_labels(labels, width: int = 16):
    return ["\n".join(textwrap.wrap(str(label), width=width)) for label in labels]


def add_bar_labels(ax, bars, values, offset: float = 0.01, fontsize: int = 9):
    for bar, value in zip(bars, values):
        label = f"{value:.3f}" if isinstance(value, (float, np.floating)) else str(value)
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + offset,
            label,
            ha="center",
            va="bottom",
            fontsize=fontsize,
            fontweight="semibold"
        )


def build_before_after_model_scores(all_subset_results: Dict[str, Dict[str, dict]]) -> pd.DataFrame:
    raw_results = all_subset_results.get("Raw Features", {})
    selected_results = all_subset_results.get("Selected Features", {})

    rows = []
    for model_name in raw_results.keys():
        raw_res = raw_results.get(model_name)
        selected_res = selected_results.get(model_name)
        if raw_res is None or selected_res is None:
            continue

        rows.append({
            "Model": model_name,
            "Before ROC-AUC": raw_res["Nested CV Mean ROC-AUC"],
            "After ROC-AUC": selected_res["Nested CV Mean ROC-AUC"],
            "Before Accuracy": raw_res["Nested CV Mean Accuracy"],
            "After Accuracy": selected_res["Nested CV Mean Accuracy"],
            "Before Precision": raw_res["Nested CV Mean Precision"],
            "After Precision": selected_res["Nested CV Mean Precision"],
            "Before Recall": raw_res["Nested CV Mean Recall"],
            "After Recall": selected_res["Nested CV Mean Recall"],
            "Before F1-score": raw_res["Nested CV Mean F1-score"],
            "After F1-score": selected_res["Nested CV Mean F1-score"],
        })

    return pd.DataFrame(rows)


def plot_metric_before_after(model_scores_df: pd.DataFrame, metric_name: str, filename: str):
    if model_scores_df.empty:
        return

    before_col = f"Before {metric_name}"
    after_col = f"After {metric_name}"
    plot_df = model_scores_df.sort_values(after_col, ascending=False).copy()
    x = np.arange(len(plot_df))
    width = 0.36

    fig, ax = plt.subplots(figsize=(13, 7))
    before_bars = ax.bar(x - width / 2, plot_df[before_col], width, label="Before", color="#8d99ae")
    after_bars = ax.bar(x + width / 2, plot_df[after_col], width, label="After", color="#2a9d8f")

    style_axis(ax, f"{metric_name} Before vs After Feature Engineering", metric_name, "Models")
    ax.set_xticks(x)
    ax.set_xticklabels(wrap_tick_labels(plot_df["Model"]), rotation=20, ha="right")
    ax.set_ylim(0, min(1.0, max(plot_df[before_col].max(), plot_df[after_col].max()) + 0.1))
    ax.legend(frameon=False, fontsize=11, loc="upper right")

    add_bar_labels(ax, before_bars, plot_df[before_col].tolist(), offset=0.008, fontsize=8)
    add_bar_labels(ax, after_bars, plot_df[after_col].tolist(), offset=0.008, fontsize=8)

    save_plot(fig, filename)


def get_representative_best_params(model_results: Dict[str, dict]) -> Dict[str, object]:
    if not model_results or "Fold Results" not in model_results:
        return {}

    params_series = model_results["Fold Results"]["Best Params"].dropna()
    if params_series.empty:
        return {}

    counts = {}
    for params in params_series:
        key = tuple(sorted(params.items()))
        counts[key] = counts.get(key, 0) + 1

    best_key = max(counts.items(), key=lambda item: item[1])[0]
    return dict(best_key)


def build_lasso_price_sensitivity_curve(
    X: pd.DataFrame,
    y: pd.Series,
    selected_features: List[str],
    all_subset_results: Dict[str, Dict[str, dict]],
    price_col: str
) -> Optional[pd.DataFrame]:
    if price_col not in X.columns or price_col not in selected_features:
        return None

    lasso_results = all_subset_results.get("Selected Features", {}).get("Lasso Logistic Regression")
    best_params = get_representative_best_params(lasso_results)

    lasso = LogisticRegression(
        penalty="l1",
        solver="liblinear",
        random_state=RANDOM_STATE,
        class_weight="balanced",
        max_iter=5000,
        C=best_params.get("classifier__C", 1.0)
    )

    pipe = build_pipeline_for_feature_subset(selected_features, X, lasso)
    pipe.fit(X[selected_features], y)

    scenario = {}
    for col in selected_features:
        if pd.api.types.is_numeric_dtype(X[col]):
            scenario[col] = float(pd.to_numeric(X[col], errors="coerce").median())
        else:
            mode = X[col].mode(dropna=True)
            scenario[col] = mode.iloc[0] if not mode.empty else "missing"

    price_values = pd.Series(pd.to_numeric(X[price_col], errors="coerce")).dropna()
    if price_values.empty:
        return None

    price_grid = np.linspace(price_values.quantile(0.05), price_values.quantile(0.95), 60)
    user_value = float(scenario.get(USER_COLUMN, pd.to_numeric(X[USER_COLUMN], errors="coerce").median())) if USER_COLUMN in X.columns else np.nan

    rows = []
    for price in price_grid:
        row = scenario.copy()
        row[price_col] = float(price)

        if "Price_Log" in selected_features:
            row["Price_Log"] = np.log1p(max(price, 0))
        if "Price_Per_User" in selected_features and pd.notna(user_value) and user_value > 0:
            row["Price_Per_User"] = price / user_value
        if "Price_Per_User_Log" in selected_features and pd.notna(user_value) and user_value > 0:
            row["Price_Per_User_Log"] = np.log1p(max(price / user_value, 0))

        row_df = pd.DataFrame([row])[selected_features]
        win_probability = pipe.predict_proba(row_df)[0, 1]
        rows.append({
            "Price": price,
            "Win_Probability": win_probability
        })

    return pd.DataFrame(rows)


def plot_lasso_price_sensitivity(
    X: pd.DataFrame,
    y: pd.Series,
    selected_features: List[str],
    all_subset_results: Dict[str, Dict[str, dict]],
    price_col: str
):
    sensitivity_df = build_lasso_price_sensitivity_curve(
        X=X,
        y=y,
        selected_features=selected_features,
        all_subset_results=all_subset_results,
        price_col=price_col
    )

    if sensitivity_df is None or sensitivity_df.empty:
        print("Lasso price sensitivity plot skipped: price column not available.")
        return

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(sensitivity_df["Price"], sensitivity_df["Win_Probability"], color="#1d3557", linewidth=2.8)
    ax.fill_between(sensitivity_df["Price"], sensitivity_df["Win_Probability"], color="#a8dadc", alpha=0.35)
    style_axis(ax, "Lasso Logistic Regression Price Sensitivity", "Predicted Win Probability", "Price")
    ax.set_ylim(0, 1)

    mid_idx = len(sensitivity_df) // 2
    mid_row = sensitivity_df.iloc[mid_idx]
    ax.scatter(mid_row["Price"], mid_row["Win_Probability"], color="#e63946", s=60, zorder=3)
    ax.annotate(
        f"Mid price\n{mid_row['Price']:.0f}\nP(win)={mid_row['Win_Probability']:.2f}",
        xy=(mid_row["Price"], mid_row["Win_Probability"]),
        xytext=(15, 15),
        textcoords="offset points",
        fontsize=10,
        bbox={"boxstyle": "round,pad=0.3", "fc": "white", "ec": "#cccccc"}
    )

    save_plot(fig, "lasso_price_sensitivity")


def build_lasso_price_probability_model(
    X: pd.DataFrame,
    y: pd.Series,
    selected_features: List[str],
    all_subset_results: Dict[str, Dict[str, dict]],
    price_col: str
):
    if price_col not in X.columns or price_col not in selected_features:
        return None, None

    lasso_results = all_subset_results.get("Selected Features", {}).get("Lasso Logistic Regression")
    best_params = get_representative_best_params(lasso_results)

    lasso = LogisticRegression(
        penalty="l1",
        solver="liblinear",
        random_state=RANDOM_STATE,
        class_weight="balanced",
        max_iter=5000,
        C=best_params.get("classifier__C", 1.0)
    )

    pipe = build_pipeline_for_feature_subset(selected_features, X, lasso)
    pipe.fit(X[selected_features], y)

    scenario = {}
    for col in selected_features:
        if pd.api.types.is_numeric_dtype(X[col]):
            scenario[col] = float(pd.to_numeric(X[col], errors="coerce").median())
        else:
            mode = X[col].mode(dropna=True)
            scenario[col] = mode.iloc[0] if not mode.empty else "missing"

    return pipe, scenario


def find_company_subset(X: pd.DataFrame, company_name: str) -> pd.DataFrame:
    if "Company" not in X.columns:
        return pd.DataFrame()

    company_series = X["Company"].astype(str)
    exact_mask = company_series.str.strip().str.lower() == company_name.strip().lower()
    if exact_mask.any():
        return X.loc[exact_mask].copy()

    contains_mask = company_series.str.lower().str.contains(company_name.strip().lower(), na=False)
    if contains_mask.any():
        return X.loc[contains_mask].copy()

    return pd.DataFrame()


def build_company_specific_scenario(
    X: pd.DataFrame,
    selected_features: List[str],
    base_scenario: Dict[str, object],
    company_name: str
) -> Optional[Dict[str, object]]:
    company_df = find_company_subset(X, company_name)
    if company_df.empty:
        return None

    scenario = base_scenario.copy()
    for col in selected_features:
        if col not in company_df.columns:
            continue
        if pd.api.types.is_numeric_dtype(company_df[col]):
            scenario[col] = float(pd.to_numeric(company_df[col], errors="coerce").median())
        else:
            mode = company_df[col].mode(dropna=True)
            if not mode.empty:
                scenario[col] = mode.iloc[0]

    scenario["Company"] = company_df["Company"].astype(str).mode().iloc[0]
    return scenario


def prompt_for_feature_value(feature_name: str, X: pd.DataFrame, default_value):
    if pd.api.types.is_numeric_dtype(X[feature_name]):
        prompt = f"{feature_name} gir [{default_value}]: "
        while True:
            user_input = input(prompt).strip()
            if user_input == "":
                return float(default_value)
            try:
                return float(user_input.replace(",", "."))
            except ValueError:
                print("Gecersiz sayisal deger. Tekrar dene.")

    unique_values = X[feature_name].dropna().astype(str).value_counts().index.tolist()
    shown_values = unique_values[:10]
    options_text = ", ".join(shown_values)
    prompt = f"{feature_name} gir [{default_value}] (ornekler: {options_text}): "
    user_input = input(prompt).strip()
    return user_input if user_input != "" else default_value


def prompt_for_full_scenario(
    X: pd.DataFrame,
    selected_features: List[str],
    scenario: Dict[str, object],
    price_col: str
) -> Dict[str, object]:
    user_scenario = scenario.copy()

    print("\nTeklif senaryosunu gir. Bos birakirsan varsayilan deger kullanilir.")
    print("Cok uzun alanlarda sana ornek degerler gosterecegim.")

    priority_features = []
    for feature in [
        "Company",
        OPPORTUNITY_TYPE_COLUMN,
        "Product",
        price_col,
        USER_COLUMN,
        "Sector",
        "Opportunity Source",
        "IAM, Cybersecurity or IT Department",
        "Company Size",
        "Competition",
        "Partner",
        "Estimated Project End Quarter",
    ]:
        if feature in selected_features and feature not in priority_features:
            priority_features.append(feature)

    remaining_features = [f for f in selected_features if f not in priority_features]
    prompt_features = priority_features + remaining_features

    auto_features = {
        "Price_Log", "Price_Per_User", "Price_Per_User_Log",
        "Offer_Year", "Offer_Month", "Offer_Quarter",
        "User_Log", "Has_Competition", "Competition_Count",
        "Has_Partner", "Product_Opportunity_Type", "Sector_Source",
        "Project_End_Known"
    }

    for feature in prompt_features:
        if feature in auto_features:
            continue
        if feature not in X.columns:
            continue
        user_scenario[feature] = prompt_for_feature_value(feature, X, scenario.get(feature))

    return user_scenario


def predict_win_probability_for_price(
    model,
    scenario: Dict[str, object],
    selected_features: List[str],
    price_col: str,
    input_price: float
) -> float:
    row = scenario.copy()
    row[price_col] = float(input_price)

    user_value = row.get(USER_COLUMN)
    if "Price_Log" in selected_features:
        row["Price_Log"] = np.log1p(max(input_price, 0))
    if "Price_Per_User" in selected_features and pd.notna(user_value) and float(user_value) > 0:
        row["Price_Per_User"] = input_price / float(user_value)
    if "Price_Per_User_Log" in selected_features and pd.notna(user_value) and float(user_value) > 0:
        row["Price_Per_User_Log"] = np.log1p(max(input_price / float(user_value), 0))

    row_df = pd.DataFrame([row])[selected_features]
    return float(model.predict_proba(row_df)[0, 1])


def plot_selected_company_sensitivity_from_scenario(
    X: pd.DataFrame,
    y: pd.Series,
    selected_features: List[str],
    all_subset_results: Dict[str, Dict[str, dict]],
    price_col: str,
    user_scenario: Dict[str, object]
):
    model, _ = build_lasso_price_probability_model(
        X=X,
        y=y,
        selected_features=selected_features,
        all_subset_results=all_subset_results,
        price_col=price_col
    )
    if model is None:
        return

    company_value = str(user_scenario.get("Company", "")).strip()
    if not company_value:
        return

    price_values = pd.Series(pd.to_numeric(X[price_col], errors="coerce")).dropna()
    if price_values.empty:
        return

    price_grid = np.linspace(price_values.quantile(0.05), price_values.quantile(0.95), 60)
    rows = []
    for price in price_grid:
        probability = predict_win_probability_for_price(
            model=model,
            scenario=user_scenario,
            selected_features=selected_features,
            price_col=price_col,
            input_price=float(price)
        )
        rows.append({"Price": float(price), "Win_Probability": probability})

    sensitivity_df = pd.DataFrame(rows)
    safe_name = "".join(ch if ch.isalnum() else "_" for ch in company_value)[:60].strip("_")

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(sensitivity_df["Price"], sensitivity_df["Win_Probability"], color="#6a4c93", linewidth=2.8)
    ax.fill_between(sensitivity_df["Price"], sensitivity_df["Win_Probability"], color="#cdb4db", alpha=0.35)
    style_axis(
        ax,
        f"Selected Company Price Sensitivity: {company_value}",
        "Predicted Win Probability",
        "Price"
    )
    ax.set_ylim(0, 1)

    current_price = float(user_scenario[price_col])
    current_probability = predict_win_probability_for_price(
        model=model,
        scenario=user_scenario,
        selected_features=selected_features,
        price_col=price_col,
        input_price=current_price
    )
    ax.scatter(current_price, current_probability, color="#d62828", s=70, zorder=3)
    ax.annotate(
        f"Selected scenario\nPrice={current_price:.0f}\nP(win)={current_probability:.2f}",
        xy=(current_price, current_probability),
        xytext=(15, 15),
        textcoords="offset points",
        fontsize=9,
        bbox={"boxstyle": "round,pad=0.3", "fc": "white", "ec": "#cccccc"}
    )

    save_plot(fig, f"lasso_price_sensitivity_selected_{safe_name}")
    print(f"Secilen firma icin grafik olusturuldu: {company_value}")


def run_lasso_price_probability_prompt(
    X: pd.DataFrame,
    y: pd.Series,
    selected_features: List[str],
    all_subset_results: Dict[str, Dict[str, dict]],
    price_col: str
):
    model, scenario = build_lasso_price_probability_model(
        X=X,
        y=y,
        selected_features=selected_features,
        all_subset_results=all_subset_results,
        price_col=price_col
    )

    if model is None or scenario is None:
        print("\nPrice probability prompt skipped: price column not available in selected features.")
        return

    print("\n" + "=" * 90)
    print("LASSO PRICE PROBABILITY COMMAND")
    print("=" * 90)
    print("Teklif bilgilerini gir. Model sana tahmini kazanma olasiligini versin.")
    print("Cikmak icin 'q' yaz.")

    while True:
        user_input = input("\nYeni bir teklif senaryosu baslatmak icin Enter, cikmak icin q: ").strip()
        if user_input.lower() in {"q", "quit", "exit"}:
            print("Price probability prompt kapatildi.")
            break

        user_scenario = prompt_for_full_scenario(
            X=X,
            selected_features=selected_features,
            scenario=scenario,
            price_col=price_col
        )
        input_price = float(user_scenario[price_col])

        probability = predict_win_probability_for_price(
            model=model,
            scenario=user_scenario,
            selected_features=selected_features,
            price_col=price_col,
            input_price=input_price
        )
        company_value = user_scenario.get("Company", "Unknown Company")
        product_value = user_scenario.get("Product", "Unknown Product")
        opp_value = user_scenario.get(OPPORTUNITY_TYPE_COLUMN, "Unknown Opportunity Type")
        print(
            f"\nSirket: {company_value}\n"
            f"Product: {product_value}\n"
            f"Opportunity Type: {opp_value}\n"
            f"Fiyat = {input_price:,.2f} icin tahmini kazanma olasiligi: %{probability * 100:.2f}"
        )
        plot_selected_company_sensitivity_from_scenario(
            X=X,
            y=y,
            selected_features=selected_features,
            all_subset_results=all_subset_results,
            price_col=price_col,
            user_scenario=user_scenario
        )


def run_company_specific_lasso_plot_prompt(
    X: pd.DataFrame,
    y: pd.Series,
    selected_features: List[str],
    all_subset_results: Dict[str, Dict[str, dict]],
    price_col: str
):
    model, base_scenario = build_lasso_price_probability_model(
        X=X,
        y=y,
        selected_features=selected_features,
        all_subset_results=all_subset_results,
        price_col=price_col
    )
    if model is None or base_scenario is None:
        print("\nCompany-specific plot skipped: price column not available.")
        return

    print("\n" + "=" * 90)
    print("COMPANY-SPECIFIC PRICE SENSITIVITY GRAPH")
    print("=" * 90)
    print("Firma adini gir. O firmaya ozel price sensitivity grafigi olusturulsun.")
    print("Cikmak icin 'q' yaz.")

    while True:
        company_name = input("\nFirma adi gir: ").strip()
        if company_name.lower() in {"q", "quit", "exit"}:
            print("Company-specific graph prompt kapatildi.")
            break
        if company_name == "":
            print("Lutfen bir firma adi gir.")
            continue

        company_scenario = build_company_specific_scenario(
            X=X,
            selected_features=selected_features,
            base_scenario=base_scenario,
            company_name=company_name
        )
        if company_scenario is None:
            print(f"Firma bulunamadi: {company_name}")
            continue

        price_values = pd.Series(pd.to_numeric(X[price_col], errors="coerce")).dropna()
        if price_values.empty:
            print("Gecerli fiyat verisi bulunamadi.")
            return

        price_grid = np.linspace(price_values.quantile(0.05), price_values.quantile(0.95), 60)
        rows = []
        for price in price_grid:
            probability = predict_win_probability_for_price(
                model=model,
                scenario=company_scenario,
                selected_features=selected_features,
                price_col=price_col,
                input_price=float(price)
            )
            rows.append({"Price": float(price), "Win_Probability": probability})

        sensitivity_df = pd.DataFrame(rows)
        safe_name = "".join(ch if ch.isalnum() else "_" for ch in str(company_scenario["Company"]))[:60].strip("_")

        fig, ax = plt.subplots(figsize=(10, 6))
        ax.plot(sensitivity_df["Price"], sensitivity_df["Win_Probability"], color="#0b6e4f", linewidth=2.8)
        ax.fill_between(sensitivity_df["Price"], sensitivity_df["Win_Probability"], color="#b7e4c7", alpha=0.4)
        style_axis(
            ax,
            f"Company Price Sensitivity: {company_scenario['Company']}",
            "Predicted Win Probability",
            "Price"
        )
        ax.set_ylim(0, 1)

        mid_idx = len(sensitivity_df) // 2
        mid_row = sensitivity_df.iloc[mid_idx]
        ax.scatter(mid_row["Price"], mid_row["Win_Probability"], color="#bc4749", s=60, zorder=3)
        ax.annotate(
            f"{company_scenario['Company']}\nPrice={mid_row['Price']:.0f}\nP(win)={mid_row['Win_Probability']:.2f}",
            xy=(mid_row["Price"], mid_row["Win_Probability"]),
            xytext=(15, 15),
            textcoords="offset points",
            fontsize=9,
            bbox={"boxstyle": "round,pad=0.3", "fc": "white", "ec": "#cccccc"}
        )

        save_plot(fig, f"lasso_price_sensitivity_{safe_name}")
        print(f"Firma ozel grafik olusturuldu: {company_scenario['Company']}")


def generate_report_plots(all_subset_results, X, y, selected_features, price_col):
    model_scores_df = build_before_after_model_scores(all_subset_results)
    plot_metric_before_after(model_scores_df, "ROC-AUC", "model_scores_before_after_rocauc")
    plot_metric_before_after(model_scores_df, "Accuracy", "model_scores_before_after_accuracy")
    plot_metric_before_after(model_scores_df, "Precision", "model_scores_before_after_precision")
    plot_metric_before_after(model_scores_df, "Recall", "model_scores_before_after_recall")
    plot_metric_before_after(model_scores_df, "F1-score", "model_scores_before_after_f1")
    plot_lasso_price_sensitivity(X, y, selected_features, all_subset_results, price_col)


# ==============================
# MAIN EXECUTION
# ==============================
data = load_dataset(
    excel_file=EXCEL_FILE,
    use_feature_engineering=True,
    drop_leakage_risk=True
)

X = data["X"]
y = data["y"]

engineered_df, removed_df = build_engineered_feature_table(data)
print("\n=== URETILEN FEATURE'LAR ===")
display(engineered_df)

print("\n=== LEAKAGE DIYE CIKARILAN KOLONLAR ===")
display(removed_df)

feature_impact_df = evaluate_feature_additions(data)
print("\n=== ENGINEERED FEATURE ROC-AUC ETKISI ===")
display(feature_impact_df)

final_features, final_features_df, dropped_engineered_df = build_final_feature_set(data, feature_impact_df)
print("\n=== FINAL KALAN FEATURE'LAR ===")
display(final_features_df)

print("\n=== GEREKSIZ BULUNAN ENGINEERED FEATURE'LAR ===")
display(dropped_engineered_df)

print("\nFinal feature list:")
print(final_features)

all_subset_results = run_feature_subset_nested_cv_comparison(
    X,
    y,
    data_dict=data,
    selected_features=final_features
)

for subset_name, subset_results in all_subset_results.items():
    print("\n" + "#" * 100)
    print(f"FEATURE SUBSET: {subset_name}")
    print("#" * 100)
    print_nested_feature_subset_results(subset_results)

final_comparison_df = combine_all_nested_subset_results(all_subset_results)

print("\n" + "=" * 100)
print("FINAL COMBINED NESTED CV COMPARISON")
print("=" * 100)
print(final_comparison_df.to_string(index=False))

save_dataframe(engineered_df, "engineered_features_table")
save_dataframe(removed_df, "removed_leakage_columns")
save_dataframe(feature_impact_df, "engineered_feature_impact_on_rocauc")
save_dataframe(final_features_df, "final_selected_features")
save_dataframe(dropped_engineered_df, "dropped_engineered_features")
save_dataframe(final_comparison_df, "all_feature_subset_nested_cv_comparison")
generate_report_plots(
    all_subset_results,
    X,
    y,
    final_features,
    data["price_col"]
)
run_lasso_price_probability_prompt(
    X,
    y,
    final_features,
    all_subset_results,
    data["price_col"]
)
run_company_specific_lasso_plot_prompt(
    X,
    y,
    final_features,
    all_subset_results,
    data["price_col"]
)
```
